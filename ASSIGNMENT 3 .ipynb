# What do you think about the students' parental level of education?

```python
df['parental level of education'].value_counts()
```

```python
plt.title('Parentel level of Education',fontsize = 20)
df['parental level of education'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.title('Gender and Parental level of Education',fontsize=20)
sns.countplot(x="parental level of education", hue="gender", data=df)
plt.show()
```

# Conclusion:
* From the pie diagram we get that 22.6% parentes were completed college education. Only 5.9% parents were studied master's degree.
* The parents of male students who completed high school education was comparatively higher than the female students,but all the remaining cases parents of female students are dominating than male students.

# Who scores the most on average for math, reading and writing based on
● Gender ● Test preparation course

```python
df[['test preparation course',
      'gender',
      'math score',
      'writing score',
      'reading score']].groupby(['test preparation course','gender']).mean()
```



# Conclusion:

# What do you think about the scoring variation for math, reading and writing based on 
● Gender ● Test preparation course

```python
df[['test preparation course','gender','math score','writing score',
      'reading score']].groupby(['test preparation course','gender']).mean()
```

```python
df[['test preparation course','gender','math score','writing score',
      'reading score']].groupby(['test preparation course','gender']).var()
```

```python
df[['test preparation course','gender','math score','writing score',
      'reading score']].groupby(['test preparation course','gender']).std()
```

# Conclusion:
   * Female category under 'none' from the test prepartion course has comparatively higer variance in scores than   the other cateogries.
   * writing score of female category under 'completed' from the test prepartion course has comparatively lower variance than the other cateogries.
